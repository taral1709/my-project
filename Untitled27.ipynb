```python
# 1. Install required library for reading PDFs
!pip install PyPDF2 textblob

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import PyPDF2
from textblob import TextBlob
import warnings
warnings.filterwarnings('ignore')

# ---------------------------------------------------------
# 2. DATA LOADING & CLEANING
# ---------------------------------------------------------
df = pd.read_csv('dataset.csv' , encoding='latin')

# Clean commas and weird characters in numbers
cols_to_clean = ['Budget', 'Revenue', 'Total_Words', 'Total_Dialogues', 'Positive_Emotion', 'Negative_Emotion', 'Overall_Vibe']
for col in cols_to_clean:
    if df[col].dtype == 'object':
        df[col] = df[col].str.replace(',', '', regex=False).str.extract(r'([\d\.]+)').astype(float)
    else:
        df[col] = pd.to_numeric(df[col], errors='coerce')

df = df.dropna(subset=['Revenue', 'Budget'])
df = df.fillna(df.median(numeric_only=True))
df['Primary_Genre'] = df['Genre'].astype(str).str.split(',').str[0].fillna('Unknown')
df['Director'] = df['Director'].astype(str).fillna('Unknown')

# ---------------------------------------------------------
# 3. ADVANCED FEATURE ENGINEERING (Target Encoding)
# ---------------------------------------------------------
# Calculate historical ROI for the dataset
df['ROI_Percentage'] = ((df['Revenue'] - df['Budget']) / df['Budget']) * 100
median_budget = df['Budget'].median()

# Target Encoding: Teach the model how successful a Director/Genre historically is
global_mean = df['Revenue'].mean()
director_means = df.groupby('Director')['Revenue'].mean()
genre_means = df.groupby('Primary_Genre')['Revenue'].mean()

df['Director_Encoded'] = df['Director'].map(director_means).fillna(global_mean)
df['Genre_Encoded'] = df['Primary_Genre'].map(genre_means).fillna(global_mean)

# ---------------------------------------------------------
# 4. MODEL TRAINING (Keeping Blockbusters)
# ---------------------------------------------------------
features = ['Total_Dialogues', 'Total_Words', 'Positive_Emotion', 'Negative_Emotion',
            'Overall_Vibe', 'Budget', 'Is_Sequel', 'Holiday_Release', 'Is_A_Rated', 'Runtime',
            'Genre_Encoded', 'Director_Encoded']

# We DO NOT remove outliers this time so the model can predict massive hits!
X = df[features]
y = df['Revenue']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Random Forest with deeper trees to capture complex script combinations
model = RandomForestRegressor(n_estimators=300, max_depth=10, random_state=42)
model.fit(X_train, y_train)

# Evaluation
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"--- MODEL EVALUATION ---")
print(f"R² Score: {r2_score(y_test, y_pred)*100:.2f}%")
print(f"Mean Absolute Error (MAE): {mae:.2f} Cr")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f} Cr\n")

# ---------------------------------------------------------
# 5. TESTING FUNCTION (With Your Custom Parameters)
# ---------------------------------------------------------
def predict_movie_success_from_pdf(pdf_path, genre, director, budget_input, is_sequel, holiday_release, is_a_rated, runtime):
    print(f"--- Analyzing Script: {pdf_path} ---")
    text = ""
    try:
        reader = PyPDF2.PdfReader(pdf_path)
        for page in reader.pages:
            text += page.extract_text() + " "
    except Exception as e:
        print(f"Error reading {pdf_path}. Make sure it is uploaded in Colab.")
        return

    # NLP text extraction
    words = text.split()
    total_words = len(words)
    total_dialogues = text.count('\n') // 2
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity
    pos_emo = max(0, polarity)
    neg_emo = abs(min(0, polarity))
    overall_vibe = 1 if polarity > 0 else 0

    # Process custom inputs dynamically
    gen_str = genre.split(',')[0]
    gen_enc = genre_means.get(gen_str, global_mean)
    dir_enc = director_means.get(director, global_mean)

    input_data = pd.DataFrame([[total_dialogues, total_words, pos_emo, neg_emo, overall_vibe,
                                budget_input, is_sequel, holiday_release, is_a_rated, runtime, gen_enc, dir_enc]],
                              columns=features)

    # Generate predictions
    pred_revenue = model.predict(input_data)[0]
    pred_roi = ((pred_revenue - budget_input) / budget_input) * 100
    risk = "High Risk" if budget_input > median_budget else "Low Risk"

    rev_min = max(0, pred_revenue - mae)
    rev_max = pred_revenue + mae

    print(f"Script Extracted -> Words: {total_words}, Dialogues: {total_dialogues}, Vibe Score: {polarity:.2f}")
    print(f"💸 Predicted Revenue: {pred_revenue:.2f} Cr")
    print(f"🎯 Revenue Range: {rev_min:.2f} Cr - {rev_max:.2f} Cr")
    print(f"📈 Estimated ROI: {pred_roi:.2f}%")
    print(f"⚠️ Risk Level: {risk}\n")

# ---------------------------------------------------------
# 6. RUN YOUR TESTS!
# ---------------------------------------------------------
# Your Exact Test for PK
predict_movie_success_from_pdf(
    pdf_path="PK.pdf",
    genre="Comedy,Drama,Musical",
    director="Rajkumar Hirani",
    budget_input=122,
    is_sequel=0,
    holiday_release=1,
    is_a_rated=0,
    runtime=133
)

# Test for Dangal
predict_movie_success_from_pdf(
    pdf_path="Dangal.pdf",
    genre="Biography,Drama,Sport",
    director="Nitesh Tiwari",
    budget_input=70,
    is_sequel=0,
    holiday_release=1,
    is_a_rated=0,
    runtime=161
)
```
